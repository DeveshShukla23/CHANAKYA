#### ============================================
#### CUSTOMERS DATA - 1000 Real Indian Customers
#### ============================================

```python
def create_customers(n=1000):
    
    # Indian cities with tier classification
    cities = {
        'Tier 1': ['Mumbai', 'Delhi', 'Bangalore', 'Chennai', 'Hyderabad', 'Kolkata', 'Pune', 'Ahmedabad'],
        'Tier 2': ['Jaipur', 'Lucknow', 'Kanpur', 'Nagpur', 'Indore', 'Bhopal', 'Patna', 'Vadodara', 'Surat', 'Agra'],
        'Tier 3': ['Jhansi', 'Meerut', 'Gwalior', 'Allahabad', 'Varanasi', 'Bareilly', 'Aligarh', 'Moradabad', 'Firozabad', 'Mathura']
    }
    
    city_tier_map = {}
    all_cities = []
    for tier, city_list in cities.items():
        for city in city_list:
            city_tier_map[city] = tier
            all_cities.append(city)
    
    # City weights — Tier 1 more customers
    city_weights = []
    for city in all_cities:
        if city_tier_map[city] == 'Tier 1': city_weights.append(5)
        elif city_tier_map[city] == 'Tier 2': city_weights.append(3)
        else: city_weights.append(1)
    
    customers = []
    for i in range(1, n+1):
        city = random.choices(all_cities, weights=city_weights)[0]
        age = random.choices(
            range(18, 65),
            weights=[5 if 22<=a<=35 else 3 if 36<=a<=45 else 1 for a in range(18, 65)]
        )[0]
        
        signup_date = fake.date_between(
            start_date=datetime(2022, 1, 1),
            end_date=datetime(2024, 12, 31)
        )
        
        customers.append({
            'customer_id'   : f'CUST{i:04d}',
            'customer_name' : fake.name(),
            'email'         : fake.email(),
            'phone'         : fake.phone_number(),
            'age'           : age,
            'gender'        : random.choices(['Male', 'Female'], weights=[55, 45])[0],
            'city'          : city,
            'state'         : fake.state(),
            'tier'          : city_tier_map[city],
            'pincode'       : fake.postcode(),
            'signup_date'   : signup_date,
            'is_premium'    : random.choices([True, False], weights=[20, 80])[0],
        })
    
    df = pd.DataFrame(customers)
    return df

# Generate Customers
df_customers = create_customers(1000)

# Save
df_customers.to_csv(r'C:\Users\Dell\CHANAKYA\data\raw\customers.csv', index=False)

print(f" Customers dataset created!")
print(f" Total Customers: {len(df_customers)}")
print(f" Cities covered: {df_customers['city'].nunique()}")
print(f"\n Customers by Tier:")
print(df_customers['tier'].value_counts().to_string())
print(f"\n Gender Split:")
print(df_customers['gender'].value_counts().to_string())
print(f"\n Premium Customers: {df_customers['is_premium'].sum()} ({df_customers['is_premium'].mean()*100:.1f}%)")
print(f"\n Age Distribution:")
print(f"   Average Age: {df_customers['age'].mean():.1f}")
print(f"   Age Range: {df_customers['age'].min()} — {df_customers['age'].max()}")
```

#### ============================================
#### ORDERS DATA - 5000 Orders (2 Years)
#### ============================================

```python
def create_orders(df_customers, n_orders=5000):
    
    # Payment methods with weights
    payment_methods = ['UPI', 'Credit Card', 'Debit Card', 'Net Banking', 'Cash on Delivery', 'EMI']
    payment_weights = [35, 20, 15, 10, 15, 5]
    
    # Order status with weights
    order_statuses = ['Delivered', 'Cancelled', 'Returned', 'Processing', 'Shipped']
    status_weights = [70, 12, 8, 5, 5]
    
    orders = []
    for i in range(1, n_orders + 1):
        
        # Random customer
        customer = df_customers.sample(1).iloc[0]
        
        # Order date — last 2 years with seasonal peaks
        month = random.choices(
            range(1, 13),
            weights=[6, 5, 6, 7, 7, 6, 7, 8, 9, 12, 15, 12]  # Oct-Dec festive season peaks
        )[0]
        year = random.choices([2023, 2024, 2025], weights=[25, 40, 35])[0]
        
        try:
            day = random.randint(1, 28)
            order_date = datetime(year, month, day)
        except:
            order_date = datetime(year, month, 1)
        
        # Delivery date
        delivery_days = random.randint(2, 7)
        delivery_date = order_date + timedelta(days=delivery_days)
        
        # Premium customers get discounts
        if customer['is_premium']:
            discount_pct = random.choices([10, 15, 20, 25], weights=[30, 30, 25, 15])[0]
        else:
            discount_pct = random.choices([0, 5, 10], weights=[50, 30, 20])[0]
        
        status = random.choices(order_statuses, weights=status_weights)[0]
        
        orders.append({
            'order_id'          : f'ORD{i:05d}',
            'customer_id'       : customer['customer_id'],
            'order_date'        : order_date.strftime('%Y-%m-%d'),
            'delivery_date'     : delivery_date.strftime('%Y-%m-%d'),
            'payment_method'    : random.choices(payment_methods, weights=payment_weights)[0],
            'order_status'      : status,
            'discount_percent'  : discount_pct,
            'shipping_city'     : customer['city'],
            'shipping_tier'     : customer['tier'],
            'is_premium_order'  : customer['is_premium'],
        })
    
    df = pd.DataFrame(orders)
    return df

# Generate Orders
df_orders = create_orders(df_customers, 5000)

# Save
df_orders.to_csv(r'C:\Users\Dell\CHANAKYA\data\raw\orders.csv', index=False)

print(f" Orders dataset created!")
print(f" Total Orders: {len(df_orders)}")
print(f" Date Range: {df_orders['order_date'].min()} to {df_orders['order_date'].max()}")
print(f"\n Order Status:")
print(df_orders['order_status'].value_counts().to_string())
print(f"\n Payment Methods:")
print(df_orders['payment_method'].value_counts().to_string())
print(f"\n Orders by City Tier:")
print(df_orders['shipping_tier'].value_counts().to_string())
```

#### ============================================
#### ORDER ITEMS DATA - Line Items for each Order
#### ============================================

```python
def create_order_items(df_orders, df_products):
    
    products_array = df_products.to_dict('records')
    order_items = []
    item_id = 1
    
    for _, order in df_orders.iterrows():
        
        n_items = random.choices([1, 2, 3, 4], weights=[50, 30, 15, 5])[0]
        selected_products = random.sample(products_array, min(n_items, len(products_array)))
        
        for product in selected_products:
            
            quantity     = random.choices([1, 2, 3], weights=[70, 20, 10])[0]
            unit_price   = product['price']
            discount_pct = order['discount_percent']
            
            discount_amount = round((unit_price * discount_pct / 100), 2)
            final_price     = round(unit_price - discount_amount, 2)
            total_amount    = round(final_price * quantity, 2)
            revenue         = total_amount if order['order_status'] != 'Returned' else 0

            order_items.append({
                'item_id'          : f'ITEM{item_id:06d}',
                'order_id'         : order['order_id'],
                'customer_id'      : order['customer_id'],
                'product_id'       : product['product_id'],
                'product_name'     : product['product_name'],
                'category'         : product['category'],
                'quantity'         : quantity,
                'unit_price'       : unit_price,
                'discount_percent' : discount_pct,
                'discount_amount'  : discount_amount,
                'final_price'      : final_price,
                'total_amount'     : total_amount,
                'revenue'          : revenue,
                'order_date'       : order['order_date'],
                'order_status'     : order['order_status'],
            })
            item_id += 1
    
    return pd.DataFrame(order_items)

# Generate
print("⏳ Generating order items... please wait ~30 seconds")
df_order_items = create_order_items(df_orders, df_products)

# Save
df_order_items.to_csv(r'C:\Users\Dell\CHANAKYA\data\raw\order_items.csv', index=False)

total_revenue = df_order_items[df_order_items['order_status']=='Delivered']['revenue'].sum()

print(f"✅ Order Items dataset created!")
print(f"📋 Total Line Items: {len(df_order_items)}")
print(f"💰 Total Revenue: ₹{total_revenue:,.2f}")
print(f"\n📊 Revenue by Category:")
cat_rev = df_order_items[df_order_items['order_status']=='Delivered'].groupby('category')['revenue'].sum().sort_values(ascending=False)
print(cat_rev.apply(lambda x: f"₹{x:,.0f}").to_string())
print(f"\n📦 Top 5 Products by Sales:")
top_prod = df_order_items.groupby('product_name')['quantity'].sum().sort_values(ascending=False).head(5)
print(top_prod.to_string())
```

```python
import time

def simulate_realtime_orders(df_customers, df_products, n_new_orders=10):
    
    print(" LIVE — CHANAKYA Real-time Order Simulation")
    print("=" * 50)
    
    new_orders      = []
    new_order_items = []
    
    # Start from next order ID
    last_order_num  = 5000
    last_item_num   = 8752
    
    payment_methods = ['UPI', 'Credit Card', 'Debit Card', 'Net Banking', 'Cash on Delivery', 'EMI']
    payment_weights = [35, 20, 15, 10, 15, 5]
    order_statuses  = ['Delivered', 'Cancelled', 'Returned', 'Processing', 'Shipped']
    status_weights  = [70, 12, 8, 5, 5]
    
    for i in range(n_new_orders):
        
        # Simulate live order coming in
        order_num       = last_order_num + i + 1
        customer        = df_customers.sample(1).iloc[0]
        order_date      = datetime.now()
        delivery_date   = order_date + timedelta(days=random.randint(2, 7))
        discount_pct    = random.choices([0, 5, 10, 15], weights=[50, 25, 15, 10])[0]
        status          = random.choices(order_statuses, weights=status_weights)[0]
        
        order = {
            'order_id'          : f'ORD{order_num:05d}',
            'customer_id'       : customer['customer_id'],
            'order_date'        : order_date.strftime('%Y-%m-%d %H:%M:%S'),
            'delivery_date'     : delivery_date.strftime('%Y-%m-%d'),
            'payment_method'    : random.choices(payment_methods, weights=payment_weights)[0],
            'order_status'      : status,
            'discount_percent'  : discount_pct,
            'shipping_city'     : customer['city'],
            'shipping_tier'     : customer['tier'],
            'is_premium_order'  : customer['is_premium'],
        }
        new_orders.append(order)
        
        # Generate items for this order
        n_items          = random.choices([1, 2, 3], weights=[60, 30, 10])[0]
        products_array   = df_products.to_dict('records')
        selected_prods   = random.sample(products_array, n_items)
        order_total      = 0
        
        for product in selected_prods:
            last_item_num  += 1
            quantity        = random.choices([1, 2], weights=[80, 20])[0]
            unit_price      = product['price']
            discount_amount = round(unit_price * discount_pct / 100, 2)
            final_price     = round(unit_price - discount_amount, 2)
            total_amount    = round(final_price * quantity, 2)
            order_total    += total_amount
            
            new_order_items.append({
                'item_id'          : f'ITEM{last_item_num:06d}',
                'order_id'         : order['order_id'],
                'customer_id'      : customer['customer_id'],
                'product_id'       : product['product_id'],
                'product_name'     : product['product_name'],
                'category'         : product['category'],
                'quantity'         : quantity,
                'unit_price'       : unit_price,
                'discount_percent' : discount_pct,
                'discount_amount'  : discount_amount,
                'final_price'      : final_price,
                'total_amount'     : total_amount,
                'revenue'          : total_amount if status != 'Returned' else 0,
                'order_date'       : order_date.strftime('%Y-%m-%d'),
                'order_status'     : status,
            })
        
        # Live print — ek ek order aata dikh raha hai!
        print(f"   [{order_date.strftime('%H:%M:%S')}] {order['order_id']} | "
              f"Customer: {customer['customer_name'][:15]:15} | "
              f"City: {customer['city']:12} | "
              f"₹{order_total:>8,.0f} | "
              f"{status}")
        
        time.sleep(0.5)  # 0.5 second delay — live feel!
    
    print("=" * 50)
    print(f" {n_new_orders} new live orders generated!")
    
    return pd.DataFrame(new_orders), pd.DataFrame(new_order_items)

# Run Simulation
df_new_orders, df_new_items = simulate_realtime_orders(df_customers, df_products, n_new_orders=10)

# Append to existing data
df_orders_final      = pd.concat([df_orders, df_new_orders], ignore_index=True)
df_items_final       = pd.concat([df_order_items, df_new_items], ignore_index=True)

# Save updated files
df_orders_final.to_csv(r'C:\Users\Dell\CHANAKYA\data\raw\orders.csv', index=False)
df_items_final.to_csv(r'C:\Users\Dell\CHANAKYA\data\raw\order_items.csv', index=False)

print(f"\n Updated Dataset Summary:")
print(f"   Total Orders     : {len(df_orders_final):,}")
print(f"   Total Line Items : {len(df_items_final):,}")
print(f"   Total Revenue    : ₹{df_items_final[df_items_final['order_status']=='Delivered']['revenue'].sum():,.2f}")
```
