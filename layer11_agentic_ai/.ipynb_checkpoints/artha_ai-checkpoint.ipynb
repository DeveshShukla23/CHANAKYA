```python
import pandas as pd
import numpy as np
import os
from groq import Groq

# Paths
PROCESSED_PATH = r'C:\Users\Dell\CHANAKYA\data\processed'

# Load all datasets
df_products = pd.read_csv(f'{PROCESSED_PATH}/products_processed.csv')
df_customers = pd.read_csv(f'{PROCESSED_PATH}/customers_processed.csv')
df_orders = pd.read_csv(f'{PROCESSED_PATH}/orders_processed.csv')
df_order_items = pd.read_csv(f'{PROCESSED_PATH}/order_items_processed.csv')

print("=" * 55)
print("  ARTHA - Agentic AI for Data Analysis")
print("  Inspired by Kautilya's Arthashastra")
print("=" * 55)
print(f"\nDatasets loaded!")
print(f"  Products    : {len(df_products):,}")
print(f"  Customers   : {len(df_customers):,}")
print(f"  Orders      : {len(df_orders):,}")
print(f"  Order Items : {len(df_order_items):,}")
```

```python
def build_data_context():
    delivered = df_order_items[df_order_items['order_status'] == 'Delivered']
    total_revenue = delivered['revenue'].sum()
    top_category = delivered.groupby('category')['revenue'].sum().idxmax()
    top_product = delivered.groupby('product_name')['revenue'].sum().idxmax()
    context = f"""
You are ARTHA, an intelligent Data Analysis AI for CHANAKYA - AI Commerce Intelligence Platform.
Inspired by Kautilya's Arthashastra, you provide wise, data-driven insights.
You have access to the following CHANAKYA dataset:
BUSINESS OVERVIEW:
- Total Products     : {len(df_products)} Indian products across 12 categories
- Total Customers    : {len(df_customers)} customers across Tier 1, 2, 3 cities
- Total Orders       : {len(df_orders)} orders (2023-2026)
- Total Order Items  : {len(df_order_items)} line items
- Total Revenue      : Rs. {total_revenue:,.2f}
- Top Category       : {top_category}
- Top Product        : {top_product}
CUSTOMER DATA columns: {list(df_customers.columns)}
ORDERS DATA columns: {list(df_orders.columns)}
ORDER ITEMS columns: {list(df_order_items.columns)}
PRODUCTS DATA columns: {list(df_products.columns)}
SAMPLE DATA INSIGHTS:
- Delivery Rate      : 70%
- Cancellation Rate  : 12.7%
- Top Payment Method : UPI (34.9%)
- Peak Month         : November (festive season)
- Champions segment  : 129 customers generating Rs. 83.4L revenue
When answering questions:
1. Always provide specific numbers from the data
2. Give actionable business recommendations
3. Be concise but insightful
4. If asked to analyze, use the actual dataset statistics
    """
    return context

context = build_data_context()
print("Data context built successfully!")
print(f"Context length: {len(context)} characters")
```

```python
from groq import Groq
from dotenv import load_dotenv
import os

load_dotenv()

GROQ_API_KEY = os.getenv("GROQ_API_KEY")

client = Groq(api_key=GROQ_API_KEY)

conversation_history = []

def ask_artha(user_question):
    
    conversation_history.append({
        "role": "user",
        "content": user_question
    })
    
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {"role": "system", "content": context},
            *conversation_history
        ],
        temperature=0.7,
        max_tokens=1024
    )
    
    answer = response.choices[0].message.content
    
    conversation_history.append({
        "role": "assistant", 
        "content": answer
    })
    return answer
print("ARTHA engine initialized!")
print(f"API Key loaded: {'Yes' if GROQ_API_KEY else 'No'}")
print("Ready to answer questions!")
```

```python
print("=" * 55)
print("  ARTHA - First Interaction Test")
print("=" * 55)

question = "What are the top 3 business insights from CHANAKYA data?"

print(f"\nQuestion: {question}")
print("\nARTHA's Answer:")
print("-" * 55)

answer = ask_artha(question)
print(answer)
```

```python
print("=" * 55)
print("  ARTHA - Multi-turn Conversation Test")
print("=" * 55)

# Conversation reset
conversation_history.clear()

questions = [
    "Which customer segment needs immediate attention and why?",
    "What should we do specifically for that segment?",
    "How much revenue can we recover if we act on your recommendation?"
]

for i, q in enumerate(questions, 1):
    print(f"\nQ{i}: {q}")
    print("-" * 55)
    answer = ask_artha(q)
    print(f"ARTHA: {answer}")
    print()
```

```python
def get_top_products(n=5):
    delivered = df_order_items[df_order_items['order_status'] == 'Delivered']
    return delivered.groupby('product_name')['revenue'].sum().nlargest(n).reset_index()

def get_revenue_by_category():
    delivered = df_order_items[df_order_items['order_status'] == 'Delivered']
    return delivered.groupby('category')['revenue'].sum().sort_values(ascending=False).reset_index()

def get_churn_risk_customers():
    import pandas as pd
    reference_date = pd.Timestamp('2026-03-18')
    df_orders['order_date'] = pd.to_datetime(df_orders['order_date'], format='mixed')
    last_order = df_orders.groupby('customer_id')['order_date'].max().reset_index()
    last_order['days_since'] = (reference_date - last_order['order_date']).dt.days
    at_risk = last_order[last_order['days_since'] >= 180]
    return at_risk.sort_values('days_since', ascending=False).head(10)

def get_monthly_revenue():
    delivered = df_order_items[df_order_items['order_status'] == 'Delivered']
    delivered = delivered.copy()
    delivered['order_date'] = pd.to_datetime(delivered['order_date'], format='mixed')
    delivered['year_month'] = delivered['order_date'].dt.to_period('M')
    return delivered.groupby('year_month')['revenue'].sum().reset_index()

def ask_artha_with_data(user_question):
    
    # Run all tools and attach real data to context
    top_products = get_top_products(5)
    cat_revenue = get_revenue_by_category()
    churn_risk = get_churn_risk_customers()
    monthly_rev = get_monthly_revenue()
    
    data_context = f"""
REAL-TIME DATA FROM CHANAKYA:

Top 5 Products by Revenue:
{top_products.to_string(index=False)}

Revenue by Category:
{cat_revenue.to_string(index=False)}

Top 10 Churn Risk Customers (180+ days inactive):
{churn_risk.to_string(index=False)}

Monthly Revenue (last 6 months):
{monthly_rev.tail(6).to_string(index=False)}
    """
    
    conversation_history.append({
        "role": "user",
        "content": user_question + "\n\nHere is the real-time data:\n" + data_context
    })
    
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {"role": "system", "content": context},
            *conversation_history
        ],
        temperature=0.7,
        max_tokens=1024
    )
    
    answer = response.choices[0].message.content
    
    conversation_history.append({
        "role": "assistant",
        "content": answer
    })
    
    return answer

# Test with real data
conversation_history.clear()
print("=" * 55)
print("  ARTHA - Agentic Analysis with Real Data")
print("=" * 55)

question = "Analyze our business performance and give top 3 urgent recommendations"
print(f"\nQuestion: {question}")
print("-" * 55)
answer = ask_artha_with_data(question)
print(f"ARTHA: {answer}")
```
